```python
import requests
```

```python
BASE = "http://127.0.0.1:8000" 

# Health check
r = requests.get(f"{BASE}/health", timeout=5)
r.raise_for_status()
r.json()
```

```python
def predecir_nps(comentario: str, base: str = BASE):
    payload = {"comentario": comentario}
    r = requests.post(f"{base}/predict", json=payload, timeout=10)
    r.raise_for_status()  # raises if 4xx/5xx
    return r.json()
```

```python
predecir_nps("mas o menos")
```
