# OOP
- Herencia: puedes extender una clase al crear una nueva clase sobre esta. Reduce repetición
- Encapsulación: Solo ves la superifcie de la clase.
- Abstracción: Mantienes oculto los métodos de la clase
- Polimorfismo: Dos clases pueden tener el mismo método

```python
class RepeatText():
    def __init__(self, n_repeats):
        self.n_repeats  = n_repeats
    # __init__ refiere una nueva instancia en la clase RepeatText.
    # self es una referencia a la instancia actual de la clase.
    # n_repeats es un parámetro que indica cuántas veces se repetirá el texto.
    # self.n_repeats almacena el valor de n_repeats en la instancia.

    def multiply(self, text):
        print((text + ' ') * self.n_repeats)
    # multiply es un método que toma un texto y lo repite n_repeats veces.
    # Nuevo método agregado a la clase RepeatText.
```

```python
repeat_twice = RepeatText(n_repeats=2)
```

```python
repeat_twice.n_repeats
```

```python
repeat_twice.multiply('Hola')
```

## Polimorfismo
- A pesar de que lr y rf son distintas clases, el mpetodo .fit() es compartido.
- Compartir un método facilita realizar cambios en los clasficadores sin cambiar otras secciones de tu código

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
lr_clf = LinearRegression()
lr_clf.fit([[1], [2], [3]], [1, 2, 3])

rf_clf = RandomForestClassifier()
rf_clf.fit([[1], [2], [3]], [1, 2, 3])
```
